## Hypothesis testing using scipy.stats functions

Hypothesis testing is a core statistical method used to make inferences about a population based on sample data. It helps evaluate claims, test relationships, and determine whether observed results are statistically significant or due to chance. Each statistical test serves a specific purpose, depending on the type of data and research question.


#### **Key Concepts**
1. **Null Hypothesis ($H_0$)**: Assumes no effect, difference, or relationship (status quo).  
2. **Alternative Hypothesis ($H_a$)**: Contradicts $H_0$ and claims a significant effect, difference, or relationship.  
3. **Test Statistic**: A calculated value to determine whether to reject $H_0$.  
4. **P-Value**: The probability of observing the result if $H_0$ is true. Smaller p-values (e.g., < 0.05) suggest stronger evidence against $H_0$.  
5. **Significance Level ($\alpha$)**: The threshold (commonly 0.05) for rejecting $H_0$.  

#### **Statistical Test Categories**
1. **Univariate Tests**: Analyze a single variable (e.g., Chi-Square Goodness-of-Fit, Kolmogorov-Smirnov Test, One-Sample t-test).  
2. **Bivariate Tests**: Examine relationships or differences between two variables (e.g., Pearson Correlation, Chi-Square Independence, Two-Sample t-test).  
3. **Multivariate Tests**: Compare three or more groups or variables (e.g., One-Way ANOVA, Kruskal-Wallis Test).  

#### **Parametric vs. Non-Parametric Tests**
- **Parametric Tests**: Assume data follows a specific distribution (e.g., normal). Examples: t-tests, ANOVA, Pearson Correlation.  
- **Non-Parametric Tests**: Do not require distribution assumptions. Examples: Mann-Whitney U Test, Kruskal-Wallis Test, Spearman Correlation.

### **Why Statistical Tests Are Essential**
- **Data Validation**: Quantify whether claims are supported by data.  
- **Decision-Making**: Inform actionable insights with statistical rigor.  
- **Confidence**: Provide measurable evidence to minimize uncertainty.

### **1. Chi-Square Test for Goodness-of-Fit**
- **Purpose**: Tests whether the observed frequencies of categorical data match the expected frequencies according to a specified distribution.
- **When to Use**: Use this when you have discrete data and want to compare observed frequencies with expected frequencies.
- **Assumptions**:
  1. Observed and expected frequencies should be non-negative.
  2. Expected frequencies should be ≥5 for validity.
- **How to Calculate**: Use `scipy.stats.chisquare(f_obs, f_exp)`.

**Example Claim**: The proportion of males and females in the "tips" dataset is 50-50.


**Hypothesis**:
- **Null Hypothesis ($H_0$)**: The observed proportions of males and females match the expected proportions (50-50 split).
- **Alternative Hypothesis ($H_a$)**: The observed proportions of males and females do not match the expected proportions.

```python
from seaborn import load_dataset
import scipy.stats as stats

# Load dataset and define population
tips = load_dataset("tips")
sample = tips['sex']
observed = sample.value_counts().values
expected = [len(sample) / 2, len(sample) / 2]

# Perform Chi-Square Test for Goodness-of-Fit
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
print(f"Chi-Square Statistic: {chi2_stat}, P-Value: {p_value}")
```

**Observations**:
- If the p-value < 0.05, we reject $H_0$ and conclude the proportions are not equally likely.
- Otherwise, the proportions of males and females in the dataset match the expected proportions (50-50 split).

- **When to Use**: Use for continuous data to check goodness-of-fit or compare two distributions.

  1. Observations are independent.

```python
import numpy as np

data = tips['total_bill']
stat, p_value = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
print(f"KS Statistic: {stat}, P-Value: {p_value}")
```

### **6. Two-Sample t-test**
- **Purpose**: Compares the means of two related samples.
- **When to Use**: Use for categorical and numerical data.
- **Assumptions**:
  1. Both samples are normally distributed.
  2. Variances of the two groups are equal.
- **How to Calculate**: Use `scipy.stats.ttest_ind(x, y)`.

**Example Claim**: The average total bill differs between smokers and non-smokers.

**Hypothesis**:
- **Null Hypothesis ($H_0$)**: Smokers and non-smokers have the same average total bill.
- **Alternative Hypothesis ($H_a$)**: Smokers and non-smokers have different average total bills.

```python
smokers = tips[tips['smoker'] == 'Yes']['total_bill']
non_smokers = tips[tips['smoker'] == 'No']['total_bill']
stat, p_value = stats.ttest_ind(smokers, non_smokers)
print(f"T-Statistic: {stat}, P-Value: {p_value}")
```

**Observations**:
- If the p-value < 0.05, we reject $H_0$ and conclude there is a difference in average total bills.
- Otherwise, the average total bill for smokers and non-smokers is the same.

### **7. One-Way ANOVA**
- **Purpose**: Compares means of three or more samples.
- **When to Use**: Use for categorical vs numerical data with >2 groups.
- **Assumptions**:
  1. Samples follow a normal distribution.
  2. Variances are equal.
- **How to Calculate**: Use `scipy.stats.f_oneway(x1, x2, ...)`.

**Example Claim**: Average total bill differs by day of the week.

**Hypothesis**:
- **Null Hypothesis ($H_0$)**: The average total bill is the same across all days.
- **Alternative Hypothesis ($H_a$)**: The average total bill differs across days.

```python
groups = [tips[tips['day'] == day]['total_bill'] for day in tips['day'].unique()]
stat, p_value = stats.f_oneway(*groups)
print(f"F-Statistic: {stat}, P-Value: {p_value}")
```

- If the p-value < 0.05, we reject $H_0$ and conclude that average total bill differs by day.
- Otherwise, the average total bill is the same across days.
