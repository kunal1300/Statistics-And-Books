### Create a Contingency Table

* To create a contingency table, begin by identifying two categorical variables from your dataset that you wish to analyze. These variables should have distinct categories (e.g., "Sunny" and "Rainy" for weather conditions or "Yes" and "No" for a decision outcome). The goal is to explore the association between these variables by summarizing their relationship in a tabular format.

* A contingency table is a structured display that organizes data into rows and columns, where:
> - Each row corresponds to one category of the first variable.
> - Each column represents one category of the second variable.
> - The cells within the table contain the frequency or count of observations for each combination of categories.

* By constructing the contingency table, you can observe how the categories of one variable are distributed across the categories of the other. This method is essential for understanding potential relationships or patterns between the two variables. For example, in a dataset on weather and activities, you could analyze how weather conditions (variable 1) influence the likelihood of engaging in an outdoor activity (variable 2).

```python
import pandas as pd
# Define the "Play Tennis" dataset
data = {
    'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 'Sunny', 'Sunny', 'Rain', 'Sunny', 'Overcast', 'Overcast', 'Rain'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High'],
    'Wind': ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Strong'],
    'Play Tennis': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

# Convert the dataset into a pandas DataFrame
df = pd.DataFrame(data)

# Generate a contingency table (e.g., Outlook vs Play Tennis)
contingency_table = pd.crosstab(df['Outlook'], df['Play Tennis'])
print("Contingency Table (Outlook vs Play Tennis):")
print(contingency_table)
```

```python
# Generate a contingency table (e.g., Outlook vs Play Tennis)
contingency_table = pd.crosstab(df['Outlook'], df['Play Tennis'], margins = True)
print("Contingency Table (Outlook vs Play Tennis):")
print(contingency_table)
```

```python
# Convert frequencies to proportions
ct_margins = pd.crosstab(df['Outlook'], df['Play Tennis'], margins=True,normalize=True)
print("Contingency Table (Outlook vs Play Tennis)")
print(ct_margins)
```

### **Probability Analysis Using Contingency Tables**
With contingency tables, we can carry out various probability analyses, such as:




**Joint Probabilities**:
   - Joint probabilities represent the likelihood of observing a specific combination of categories (e.g., `Sunny` and `Yes`). These are obtained by dividing the frequency of the cell by the overall sample size.

```python
# Calculate Joint Probabilities
joint_probabilities = pd.crosstab(df['Outlook'], df['Play Tennis'], normalize=True)
print("\nJoint Probabilities (Outlook and Play Tennis):")
print(joint_probabilities)
```

**Observations:**
* For the given dataset, joint probabilities indicate the relative frequencies of Play Tennis outcomes (Yes or No) across different Outlook categories.
* For example, if the joint probability for Sunny and Yes is 0.25, it implies that 25% of the total observations fall into this combination.
* Higher joint probabilities for combinations like Overcast and Yes suggest a stronger association between favorable weather and playing tennis.


**Marginal Probabilities**:
   - Marginal probabilities are calculated using the totals in the rows or columns of the table. For example, the probability of observing a specific `Outlook` (e.g., `Sunny`) is determined by dividing the total frequency for that category by the overall sample size.

```python
# Find the marginal probabilities
marginal_probabilities = pd.crosstab(df['Outlook'], df['Play Tennis'], normalize=True, margins = True)
print("Contingency Table (Outlook vs Play Tennis): Relative frequencies")
print(marginal_probabilities)
```

**Observations:**
* Marginal probabilities for Outlook categories (e.g., Sunny, Rain, Overcast) reveal the overall distribution of weather types.
* Marginal probabilities for Play Tennis categories (e.g., Yes, No) show the proportion of days tennis is played or not played.
* For instance, if the marginal probability for Play Tennis = Yes is 0.65, it suggests that tennis is played on 65% of the observed days.


**Conditional Probabilities**:
   - Conditional probabilities measure the probability of one category given another. For example, the probability of `Play Tennis = Yes` given that `Outlook = Sunny` is calculated by dividing the frequency for `Sunny` and `Yes` by the total frequency for `Sunny`.

```python
print(0.285714/0.642857)
print(0.214286/0.642857)
```

```python
# Calculate Conditional Probabilities (e.g., P(Outlook | Play Tennis))
conditional_probabilities_ogpt = marginal_probabilities.apply(lambda x: x/x[-1])
print("\nConditional Probabilities (Outlook given Play Tennis):")
print(conditional_probabilities_ogpt)
```

```python
# Calculate Conditional Probabilities (e.g., P(Play Tennis | Outlook))
conditional_probabilities_ptgo = marginal_probabilities.apply(lambda x: x/x[-1], axis = 1)
print("\nConditional Probabilities (Play Tennis given Outlook):")
print(conditional_probabilities_ptgo)
```

**Observations:**
* The conditional probability of Play Tennis = Yes varies across different Outlook categories. For example:
* If P(Play Tennis = Yes | Outlook = Sunny) is lower than P(Play Tennis = Yes | Outlook = Overcast), this indicates that favorable weather (Overcast) has a higher likelihood of tennis being played.
* The conditional probabilities are computed separately for each pair of categories, helping identify dependencies between variables.


**Independence Analysis**:
* Contingency tables help test whether two categorical variables are independent. Statistical methods such as the Chi-Square test can be applied to the table for this purpose. ()
* If the conditional probabilities for  across categories of  are equal to the marginal probability for , the variables are independent.
> * **Independence:** If the conditional probabilities are equal to the marginal probabilities for all categories, the two variables are independent.
> * **Dependence:** If the conditional probabilities vary across categories, the two variables are not independent.



```python
# Comparing Conditional Probabilities to Marginal Probabilities
print("\nComparison of Conditional Probabilities (Play Tennis given Outlook) to Marginal Probabilities:")
for outlook in conditional_probabilities_ptgo.index[:-1]:
    for play_tennis in conditional_probabilities_ptgo.columns[:-1]:
        conditional_prob = conditional_probabilities_ptgo.loc[outlook, play_tennis]
        marginal_prob = marginal_probabilities.loc['All', play_tennis]
        print(f"P({play_tennis} | {outlook}) = {conditional_prob:.2f}, Marginal P({play_tennis}) = {marginal_prob:.2f}")
        print(f"Difference: {abs(conditional_prob - marginal_prob):.2f}")
```

**Observations:**
* By comparing conditional probabilities to marginal probabilities, we assess independence:
> * If conditional probabilities (e.g., P(Play Tennis = Yes | Outlook)) match marginal probabilities (e.g., P(Play Tennis = Yes)), the variables are independent.
> * Deviations suggest dependence. For instance, if Outlook = Overcast significantly increases the likelihood of Play Tennis = Yes, the two variables are not independent.
* By comparing conditional and marginal probabilities, it is evident that  and  are dependent variables. This dependency can be statistically tested further using a Chi-Square test.

**Visualization**:
   - Conditional proportions (probabilities) can be visualized using stacked bar charts or mosaic plots to better understand the relationship between the variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a stacked bar plot using seaborn's countplot
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Outlook', hue='Play Tennis', palette='viridis')
plt.title('Stacked Bar Plot of Outlook vs Play Tennis', fontsize=14)
plt.xlabel('Outlook', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Play Tennis')
plt.tight_layout()
plt.show()
```

**Stacked Bar Plot:**
* Visualizing Outlook and Play Tennis with a stacked bar plot provides an intuitive understanding of the distribution of Yes and No outcomes across different weather conditions.
* For example:
> * A prominent portion of Play Tennis = Yes under Overcast suggests its favorable impact.
> * A balanced proportion of Yes and No under Sunny highlights a neutral impact of sunny conditions.


## Tips Dataset

The `tips` dataset is a widely used example dataset in data visualization and statistical analysis, particularly in Python-based libraries like seaborn and pandas. It contains information collected from a restaurant about bills, tips, and customer demographics. This dataset is often employed to study tipping behavior and examine relationships between various features.

### **Dataset Features**
Here are the columns included in the dataset:
1. **total_bill**:
   - The total bill amount (including food and drinks) paid by the customer.
   - Numeric feature used to analyze spending patterns.

2. **tip**:
   - The amount of tip given by the customer.
   - Numeric feature used to analyze tipping behavior relative to other factors like bill size and demographics.

3. **sex**:
   - The gender of the customer (Male or Female).
   - Categorical feature used to study whether tipping behavior varies by gender.

4. **smoker**:
   - Whether the customer was a smoker (Yes or No).
   - Categorical feature used to examine the influence of smoking habits on tipping behavior.

5. **day**:
   - The day of the week when the customer visited the restaurant (e.g., Thurs, Fri, Sat, Sun).
   - Categorical feature used to analyze tipping patterns across different days.

6. **time**:
   - The time of day when the meal was served (Lunch or Dinner).
   - Categorical feature used to study tipping behavior during different meal times.

7. **size**:
   - The number of people in the dining party.
   - Numeric feature used to analyze the effect of group size on tipping.

### Import the Dataset
The  dataset can be directly accessed through the seaborn library. We will load it into a pandas DataFrame for analysis.

```python
# Load the `tips` dataset from seaborn
tips = sns.load_dataset('tips')
tips.head()
```

### Generate a Contingency Table
Construct a contingency table to summarize the frequency of combinations of sex and smoker.

```python
# Generate a Contingency Table for `sex` and `smoker`
contingency_table = pd.crosstab(tips['sex'], tips['smoker'])
print("Contingency Table (Sex vs Smoker):")
print(contingency_table)
```

**Observations**
* The contingency table summarizes the frequency of sex and smoker combinations.
* It reveals the distribution of smokers and non-smokers among males and females.


### Perform Probability Analysis
Joint Probabilities: Calculate the probability of observing specific combinations of sex and smoker.

```python
# Calculate Joint Probabilities
joint_probabilities = pd.crosstab(tips['sex'], tips['smoker'], normalize=True)
print("\nJoint Probabilities (Sex and Smoker):")
print(joint_probabilities)
```

**Observations**
* Joint probabilities measure the likelihood of observing a specific combination of sex and smoker.
* These values provide insights into how smoking behavior is distributed across genders.


Marginal Probabilities: Find the overall probabilities of each category in sex and smoker.

```python
# Find Marginal Probabilities
marginal_probabilities = pd.crosstab(tips['sex'], tips['smoker'], normalize=True, margins=True)
print("\nMarginal Probabilities (Sex and Smoker):")
print(marginal_probabilities)
```

**Observations**
* Marginal probabilities show the overall likelihood of being a smoker or non-smoker, regardless of gender.
* Similarly, the probabilities of being male or female are calculated irrespective of smoking behavior.


Conditional Probabilities: Compute the probabilities of  given  and sex given smoker.

```python
# Calculate Conditional Probabilities
conditional_probabilities_smoker_given_sex = contingency_table.div(contingency_table.sum(axis=1), axis=0)
conditional_probabilities_sex_given_smoker = contingency_table.div(contingency_table.sum(axis=0), axis=1)

print("\nConditional Probabilities (Smoker given Sex):")
print(conditional_probabilities_smoker_given_sex)

print("\nConditional Probabilities (Sex given Smoker):")
print(conditional_probabilities_sex_given_smoker)
```

**Observations**
* Smoker given Sex:
> * Conditional probabilities for smokers given gender reveal how smoking behavior varies between males and females.
* Sex given Smoker:
> * Conditional probabilities for gender given smoking behavior provide insights into gender-specific smoking tendencies.

```python
# Independence Analysis
print("\nComparison of Conditional Probabilities to Marginal Probabilities:")
for sex in conditional_probabilities_smoker_given_sex.index:
    for smoker in conditional_probabilities_smoker_given_sex.columns:
        cond_prob = conditional_probabilities_smoker_given_sex.loc[sex, smoker]
        marg_prob = marginal_probabilities.loc['All', smoker]
        print(f"P({smoker} | {sex}) = {cond_prob:.2f}, Marginal P({smoker}) = {marg_prob:.2f}")
        print(f"Difference: {abs(cond_prob - marg_prob):.2f}")
```

**Observations**
* By comparing conditional probabilities to marginal probabilities, we determine whether sex and smoker are independent variables.
* Deviations between these probabilities suggest an association between gender and smoking behavior.
