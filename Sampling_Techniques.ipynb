# Sampling Techniques

## Sampling techniques

Sampling techniques form the foundation of statistical analysis and research methodologies. These techniques provide systematic ways to select a subset (sample) from a larger group (population) for analysis. By working with a sample, researchers can infer characteristics about the entire population without the need to study every individual or item, which may be impractical due to constraints like time, cost, or feasibility.

Here’s a concise summary of the sampling techniques demonstrated:

---

### **Sampling Techniques Summary**

1. **Simple Random Sampling (SRS)**:
   - **Description**: Selects items randomly, ensuring each has an equal chance of being chosen.
   - **Methods Used**:
     - `np.choose`: Selects random indices for sampling.
     - `dataframe.sample`: A direct pandas function to randomly pick rows from the dataset.
   - **Strengths**: Highly unbiased and representative.
   - **Limitations**: Does not guarantee subgroup representation and requires a complete list of the population.

2. **Systematic Sampling**:
   - **Description**: Selects items at regular intervals from a shuffled dataset.
   - **Method Used**:
     - Shuffle the dataset using `dataframe.sample` and slice rows at steps of `n` using `iloc`.
   - **Strengths**: Easy to implement and ensures structured selection.
   - **Limitations**: May introduce bias if the population has periodic patterns.

3. **Stratified Sampling**:
   - **Description**: Divides the population into subgroups (strata) and performs random sampling within each group.
   - **Method Used**:
     - Group the dataset by categorical features (e.g., `sex`) using `groupby`, followed by random sampling from each group using `apply`.
   - **Strengths**: Guarantees representation across subgroups and reduces variability in results.
   - **Limitations**: Requires detailed knowledge of population characteristics and is more complex to execute.


### Implementation of sampling techniques

```python
import pandas as pd
import numpy as np
import seaborn as sns

# Load dataset
tips = sns.load_dataset('tips')

# Tips dataset
data = {
    'total_bill': tips.total_bill,
    'tip': tips.tip,
    'sex': tips.sex
}
df = pd.DataFrame(data)

print("Original Dataset:\n", df.head())
```

1. **Simple Random Sampling**:
   - **Using `np.choose`**: Random indices are selected, and rows are fetched using those indices.
   - **Using `dataframe.sample`**: A direct function provided by pandas to randomly sample rows. It’s simpler and widely used.

```python
# Using np.choose (indices need to be specified)
n_samples = 5
random_indices_np = np.random.choice(df.index, size=n_samples, replace=False)
simple_random_sample_np = df.loc[random_indices_np]
print("\nSimple Random Sampling using np.choose:\n", simple_random_sample_np)
```

```python
# Using dataframe.sample
simple_random_sample_df = df.sample(n=n_samples, random_state=42)
print("\nSimple Random Sampling using dataframe.sample:\n", simple_random_sample_df)
```

2. **Systematic Sampling**:
   - Introduces structure by selecting items at fixed intervals.
   - May fail if data has periodic patterns.

```python
# --- 2. Systematic Sampling ---

# Shuffle and slice in steps of n
n_step = 50
shuffled_df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle rows
systematic_sample = shuffled_df.iloc[::n_step]  # Take every nth row
print("\nSystematic Sampling using shuffle and slicing:\n", systematic_sample)
```

3. **Stratified Sampling**:
   - Guarantees subgroup representation (e.g., equal sampling from `Male` and `Female` groups).
   - More complex to implement but useful when population has distinct subgroups.

```python
# --- 3. Stratified Sampling ---

# Group by strata (sex) and perform simple random sampling within groups
strata = 'sex'
stratified_sample = df.groupby(strata, group_keys=False).apply(lambda x: x.sample(n=5, random_state=42))
print("\nStratified Sampling using groupby:\n", stratified_sample)
```
