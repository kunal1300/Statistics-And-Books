## Q9. Physical Activity Level follows Poisson distribution
**H0:** Data follows Poisson(λ)
**H1:** Data does not follow Poisson distribution


```python
import pandas as pd
import numpy as np
from scipy import stats

url = "https://raw.githubusercontent.com/Muhanad-husn/Sleep-Health-and-Lifestyle/refs/heads/main/data.csv"
df = pd.read_csv(url)

data = df['Physical Activity Level']
lam = data.mean()
obs = data.value_counts().sort_index()
exp = stats.poisson.pmf(obs.index, mu=lam) * len(data)
exp = exp * (obs.sum()/exp.sum())
chi2, p = stats.chisquare(obs, exp)
print(chi2, p)
```

## Q10. Sleep Duration Normal or Exponential Distribution
**Normal Test (Shapiro)**
**H0:** Normal distribution


```python
stat, p = stats.shapiro(df['Sleep Duration'])
print(stat, p)
```

**Exponential Test (KS Test)**


```python
data = df['Sleep Duration']
loc, scale = stats.expon.fit(data)
stat, p = stats.kstest(data, 'expon', args=(loc, scale))
print(stat, p)
```

## Q11. Relationship between Age and Sleep Duration
**H0:** No correlation


```python
sample = df.sample(n=40, random_state=42)
corr, p = stats.pearsonr(sample['Age'], sample['Sleep Duration'])
print(corr, p)
```

## Q12. Women have longer sleep duration than men


```python
sample = df.sample(n=40, random_state=72)
women = sample[sample['Gender']=='Female']['Sleep Duration']
men = sample[sample['Gender']=='Male']['Sleep Duration']
t, p2 = stats.ttest_ind(women, men)
p = p2/2 if t>0 else 1-(p2/2)
print(t, p)
```

## Q13. Women have longer sleep duration than men (log transform)


```python
sample = df.sample(n=40, random_state=221)
women = np.log(sample[sample['Gender']=='Female']['Sleep Duration'])
men = np.log(sample[sample['Gender']=='Male']['Sleep Duration'])
t, p2 = stats.ttest_ind(women, men)
p = p2/2 if t>0 else 1-(p2/2)
print(t, p)
```

## Q14. Women have lower sleep quality than men


```python
sample = df.sample(n=30, random_state=9)
women = sample[sample['Gender']=='Female']['Quality of Sleep']
men = sample[sample['Gender']=='Male']['Quality of Sleep']
t, p2 = stats.ttest_ind(women, men)
p = p2/2 if t<0 else 1-(p2/2)
print(t, p)
```

## Q15. Higher stress leads to lower sleep duration


```python
sample = df.sample(n=10, random_state=979)
corr, p2 = stats.pearsonr(sample['Stress Level'], sample['Sleep Duration'])
p = p2/2 if corr<0 else 1-(p2/2)
print(corr, p)
```

## Q16. No difference among Doctors, Engineers, Lawyers


```python
sample = df.sample(n=10, random_state=979)
subset = sample[sample['Occupation'].isin(['Doctor','Engineer','Lawyer'])]
groups = [
 subset[subset['Occupation']=='Doctor']['Sleep Duration'],
 subset[subset['Occupation']=='Engineer']['Sleep Duration'],
 subset[subset['Occupation']=='Lawyer']['Sleep Duration']
]
f, p = stats.f_oneway(*groups)
print(f, p)
```
