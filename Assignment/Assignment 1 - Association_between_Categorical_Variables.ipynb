# Q.2)  What is the probability that a randomly selected individual has a stress level of 5? 

```python
from scipy.stats import norm
mean = df["Stress Level"].mean()
std = df["Stress Level"].std()

probability = norm.cdf(5, loc=mean, scale=std)
print("Probability that stress level <= 5:", probability)
```

# Q.3)  What is the probability that a randomly selected individual has a quality of sleep score of 7?

```python
mean = df["Quality of Sleep"].mean()
std = df["Quality of Sleep"].std()

probability = norm.cdf(7, loc=mean, scale=std)
print("Probability that stress level <= 7:", probability)
```

# Q.4)  What is the probability that a randomly selected individual has a stress level at least 5? 

```python
mean = df["Stress Level"].mean()
std = df["Stress Level"].std()

probability = 1 - norm.cdf(5, loc=mean, scale=std)
print("Probability that stress level >= 5:", probability)
```
