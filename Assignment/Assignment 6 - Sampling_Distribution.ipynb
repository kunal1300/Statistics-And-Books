# Load Dataset

```python
import pandas as pd
import numpy as np
from scipy import stats
import warnings
warnings.filterwarnings("ignore")

# Load dataset
url = 'https://raw.githubusercontent.com/Muhanad-husn/Sleep-Health-and-Lifestyle/refs/heads/main/data.csv'
df = pd.read_csv(url)
df.head()
```

# Q.1) Estimate the proportion of females using a large number of samples (seed = 120). 

```python
p_female = 185 / 374
np.random.seed(120)

samples = np.random.choice([0, 1], size=100000, p=[1 - p_female, p_female])

estimated_proportion = np.mean(samples)

print("Estimated Proportion of Females:", estimated_proportion)
```

# Q.2) Calculate the standard error of the estimated proportion of females and comment on its implications. 

```python
p = 185 / 374
n = 374

std_error = (p * (1 - p) / n) ** 0.5

print("Standard Error is:", std_error)
```

# Q.3) Analyze how varying sample size and number of samples affects the estimated proportion of females and standard errors. 

```python
p = 185 / 374
np.random.seed(120)

sample_sizes = [50, 100, 200, 500, 1000]
num_simulations_list = [100, 1000, 10000]

for sims in num_simulations_list:
    print(f"\nNumber of Simulations: {sims}")
    
    for n in sample_sizes:
        estimates = []
        
        for _ in range(sims):
            sample = np.random.choice([0, 1], size=n, p=[1-p, p])
            estimates.append(np.mean(sample))
        
        est_prop = np.mean(estimates)
        se = np.std(estimates)
        
        print(f"Sample Size={n} -> Estimated p={round(est_prop,4)}, SE={round(se,4)}")
```

# Q.4) Estimate the population mean of sleep duration using a large number of samples (seed = 120). 

```python
np.random.seed(120)
num_samples = 10000
sample_size = 50

means = []
for _ in range(num_samples):
    sample = df["Sleep Duration"].sample(sample_size, replace=True)
    means.append(sample.mean())
estimated_mean = np.mean(means)

print("Estimated Mean Sleep Duration:", estimated_mean)
```

#
------

# Q.5) Calculate the standard error of the estimated population mean of sleep duration. 

```python
n = len(df["Sleep Duration"])

std = df["Sleep Duration"].std()
std_error = std / (n ** 0.5)

print("Standard Error is:",std_error)
```

# Q.6) Analyze how varying sample size and number of samples impacts the estimated population mean of sleep duration and standard errors

```python
np.random.seed(120)

sample_sizes = [30, 50, 100, 200, 500]
num_simulations_list = [100, 1000, 10000]

for sims in num_simulations_list:
    print(f"\nSimulations: {sims}")
    
    for n in sample_sizes:
        means = []
        
        for _ in range(sims):
            sample = df["Sleep Duration"].sample(n, replace=True)
            means.append(sample.mean())
        
        est_mean = np.mean(means)
        se = np.std(means)
        
        print(f"n={n} -> Mean={round(est_mean,3)}, SE={round(se,3)}")
```
