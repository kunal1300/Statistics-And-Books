```python
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
from sklearn.preprocessing import LabelEncoder
 
df = pd.read_csv("https://raw.githubusercontent.com/Muhanad-husn/Sleep-Health-and-Lifestyle/refs/heads/main/data.csv")

df.head()
```

# Q.1) Formulate null and alternative hypotheses and determine the nature of the hypothesis test: 

### Example structure (for mean tests)
##### H0: μ = 8
#### H1: μ != 8 (two-tailed)

#### For proportion:
#### H0: p >= 0.5
#### H1: p < 0.5

# Q.2) Mean sleep duration is 8 hours (population standard deviation = 1.5, 95% confidence): 

```python
mu = 8
sigma = 1.5
alpha = 0.05

sample = df['Sleep Duration']
n = len(sample)
x_bar = sample.mean()

z = (x_bar - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z-stat:", z)
print("P-value:", p_value)

if p_value < alpha:
    print("Decision: Reject the Null Hypothesis")
else:
    print("Decision: Fail to reject the Null Hypothesis")
```

# Q.3) Mean sleep duration is less than 8 hours (population standard deviation = 1.5, 95% confidence): 

```python
z = (x_bar - mu) / (sigma / np.sqrt(n))
p_value = stats.norm.cdf(z)

print("Z-stat:", z)
print("P-value:", p_value)

if p_value < alpha:
    print("Decision: Reject the Null Hypothesis")
else:
    print("Decision: Fail to reject the Null Hypothesis")
```

# Q.4) Mean sleep duration is greater than 8 hours (population standard deviation = 1.5, 95% confidence): 

```python
z = (x_bar - mu) / (sigma / np.sqrt(n))
p_value = 1 - stats.norm.cdf(z)

print("Z-stat:", z)
print("P-value:", p_value)

if p_value < alpha:
    print("Decision: Reject the Null Hypothesis")
else:
    print("Decision: Fail to reject the Null Hypothesis")
```

# Q.5) At least 50% experience sleep disorders (random state = 42, 95% confidence): • H0 : π ≥ 0.50

```python
np.random.seed(42)

sample_df = df.sample(frac=1, random_state=42)

count = (sample_df['Sleep Disorder'] != 'None').sum()
n = len(sample_df)

stat, p_value = proportions_ztest(count, n, value=0.5, alternative='smaller')

print("Z-stat:", stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Decision: Reject the Null Hypothesis")
else:
    print("Decision: Fail to reject the Null Hypothesis")
```

# Q.6) BMI categories are equally distributed (95% confidence): • H0 : πOverweight = πNormal = 0.5

```python
bmi_counts = df['BMI Category'].value_counts()

chi2, p_value = stats.chisquare(bmi_counts)

print("Chi-square:", chi2)
print("P-value:", p_value)

if p_value < alpha:
    print("Decision: Reject the Null Hypothesis")
else:
    print("Decision: Fail to reject the Null Hypothesis")
```

# Q.7) Relationship between gender and sleep disorders (sample size = 60, random state = 42, 95% confidence):

```python
np.random.seed(42)

sample_df = df.sample(n=60, random_state=42)

contingency = pd.crosstab(sample_df['Gender'], sample_df['Sleep Disorder'])

chi2, p_value, dof, expected = stats.chi2_contingency(contingency)

print("Chi-square:", chi2)
print("P-value:", p_value)

if p_value < alpha:
    print("Decision: Reject the Null Hypothesis")
else:
    print("Decision: Fail to reject the Null Hypothesis")
```

# Q.8) Relationship between sleep duration and physical activity (sample size = 30, random state = 42, 95% confidence):

```python
np.random.seed(42)

sample_df = df.sample(n=30, random_state=42)

x = sample_df['Sleep Duration']
y = sample_df['Physical Activity Level']

corr, p_value = stats.pearsonr(x, y)

print("Correlation:", corr)
print("P-value:", p_value)

if p_value < alpha:
    print("Decision: Reject the Null Hypothesis")
else:
    print("Decision: Fail to reject the Null Hypothesis")
```
